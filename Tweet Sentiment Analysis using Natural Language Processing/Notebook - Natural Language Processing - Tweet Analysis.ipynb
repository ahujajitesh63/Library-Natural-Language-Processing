```python
import tweepy
```

```python
from tweepy import OAuth1UserHandler
```

```python
consumer_key = "Enter your tweepy consumer key"
consumer_secret = "Enter your tweepy consumer secret"
access_key = "Enter your tweepy access key"
access_secret = "Enter your tweepy access secret"

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
```

```python
api = tweepy.API(auth)
```

```python
from datetime import date, timedelta
import datetime
import re
from tkinter import *
from tkcalendar import Calendar


def get_tweet(count,keyword, start_date, end_date):
    

    tweets = tweepy.Cursor(api.search_tweets,
                   q=keyword,
                   lang="en",
                   since_id = start_date,
                    until = end_date,
                   tweet_mode='extended',
                   result_type="popular"
                   ).items(count)
    
    latest_tweet = [re.sub("[^A-Za-z0-9]+"," ", tweet.full_text) for tweet in tweets]
    
    
    while not latest_tweet:
             tweets = tweepy.Cursor(api.search_tweets,
                   q=keyword,
                   lang="en",
                   since_id = start_date,
                    until = end_date,
                   tweet_mode='extended',
                   result_type="popular"
                   ).items(count)
             latest_tweet = [re.sub("[^A-Za-z0-9]+"," ", tweet.full_text) for tweet in tweets]
        
    
    
    return latest_tweet    
    
```

```python
from textblob import TextBlob

def analyze_sentiment():
    
    count = 0
    positive_count = 0
    negative_count = 0
    avg_sentiment = 0
    negative_avg_sentiment = 0
    positive_avg_sentiment = 0
    total_polarity = 0
    negative_total_polarity = 0 
    positive_total_polarity = 0
    count_tweets = input ("Enter Tweet Count to analyze: ")
    checked_tweets = 0
    keywords = input("Enter keyword - case insensitive: ")
    show_tweets = input("Display all tweets? Enter y/n: ")
    show_tweets = show_tweets.lower()
    
    date_entry = input('Enter a start_date (i.e. yyyy/mm/dd): ')
    year, month, day = map(int, date_entry.split('/'))
    start_date = date(year, month, day)


    date_entry = input('Enter a end_date (i.e. yyyy/mm/dd): ')
    year, month, day = map(int, date_entry.split('/'))
    end_date = date(year, month, day)
    
    print ("Analyzing ", keywords)
    print ("_____________________________________________")
    
    latest_tweet = get_tweet(int(count_tweets), keywords, start_date, end_date)
    for tweet in latest_tweet:
        checked_tweets = checked_tweets + 1
        tweet_blob = TextBlob(tweet.lower())
        total_polarity = total_polarity + tweet_blob.sentiment.polarity
        
        if(tweet_blob.sentiment.polarity > 0):
            count = count + 1
            positive_count = positive_count+1
            positive_total_polarity = positive_total_polarity + tweet_blob.sentiment.polarity
            positive_avg_sentiment = positive_total_polarity/positive_count
      
            
                    
        if(tweet_blob.sentiment.polarity<0) :
                    count = count + 1
                    negative_count = negative_count+1
                    negative_total_polarity = negative_total_polarity + tweet_blob.sentiment.polarity
                    negative_avg_sentiment = negative_total_polarity/negative_count

        if (count !=0):
            avg_sentiment = total_polarity/count
            
        if (show_tweets == 'y' or show_tweets == 'yes'): 
            if(tweet_blob.sentiment.polarity < 0):
                print("Tweet with Negative Sentiment:")
                print("TWEET: ", tweet) 
                print("POLARITY:", tweet_blob.sentiment.polarity)
                print("____________________________________________________")
            elif(tweet_blob.sentiment.polarity <= 0):
                print("Tweet with Neutral Sentiment:")
                print("TWEET: ", tweet) 
                print("POLARITY:", tweet_blob.sentiment.polarity)
                print("____________________________________________________")
            elif(tweet_blob.sentiment.polarity > 0):
                print("Tweet with Positive Sentiment:")
                print("TWEET: ", tweet) 
                print("POLARITY:", tweet_blob.sentiment.polarity)
                print("____________________________________________________")
                

           
#     if(checked_tweets == int(count_tweets)) :
    print ("Details of keyword ", keywords )
    print ("Only ", checked_tweets, "tweets found between given dates! ")
    print ("Average sentiment of ", checked_tweets, " tweets is:", avg_sentiment)
    print ("Total count of Negative Sentiments: ", 
                   negative_count, " Having Average Negative Sentiment: ",negative_avg_sentiment)
    print ("Total count of Positive Sentiments: ", 
                   positive_count, " Having Average Positive Sentiment: ",positive_avg_sentiment)
    print("Analysis complete!")    
            
    
analyze_sentiment() 
```
